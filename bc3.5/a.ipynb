```python
def find_split_points(numbers, M):
    if not numbers:
        return [0]  # or []? but 0 is default, so assume non-empty
    
    n = len(numbers)
    
    # 验证给定 f 是否可行
    def valid(f):
        splits = [0]
        for i in range(1, n):
            if numbers[i] > f:
                splits.append(i)
        # 检查相邻分割点间距
        for i in range(1, len(splits)):
            if splits[i] - splits[i-1] > M:
                return False
        return True

    # 二分搜索最大 f
    lo = min(numbers) - 1
    hi = max(numbers) + 1

    # 如果 f = hi, 没有任何 numbers[i] > f, 所以 splits = [0] → always valid?
    # 但若 n > 1 且 M < n-1, [0] alone is still valid because only one point → no gap to check.
    # 所以 valid(hi) 总是 True.
    # 我们要最大 f 使得 valid(f) is True → 所以答案在 [lo, hi]

    for _ in range(100):  # enough for float precision
        mid = (lo + hi) / 2
        if valid(mid):
            lo = mid  # can try larger f
        else:
            hi = mid

    best_f = lo

    # Build split points using best_f
    splits = [0]
    for i in range(1, n):
        if numbers[i] > best_f:
            splits.append(i)
    
    return splits,best_f


print(find_split_points([1, 11, 3, 4, 5, 60, 7, 8, 9, 10], 3))
```

```python
from collections import deque
def find_split_points(numbers, m):
    if m >= len(numbers): return [0]
    def sliding_window_max_indices(arr, M):
        
        N = len(arr)
        
        dq = deque()  # 存储索引
        result = []
        
        for i in range(N):
            # 移除队列中超出当前窗口范围的索引（窗口是 [i-M+1, i]）
            while dq and dq[0] < i - M + 1:
                dq.popleft()
            
            # 从队尾移除所有对应值 <= arr[i] 的索引（保持单调递减）
            while dq and arr[dq[-1]] <= arr[i]:
                dq.pop()
            
            dq.append(i)
            
            # 当窗口形成（即 i >= M - 1），记录最大值的索引
            if i >= M - 1:
                result.append(dq[0])
        
        return result

    max_pos = sliding_window_max_indices(numbers, m+1)
    
    splits = [0]
    st = 1

    while st < len(max_pos):
        split = max_pos[st]
        splits.append(split)
        st = split + 1
        
    return splits
    
    
arr = [1, 3, -1, -3, 5, 3, 6, 7]
M = 3
indices = sliding_window_max_indices(arr, M)
print("Max value indices:", indices)
```
